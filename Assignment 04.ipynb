Q4.COUNT FREQUENCY OF NUMBERS 

```python
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

freq = {}

for num in numbers:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

Q5.WORD COUNTER

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print(freq)
```

Q6.Student marks and average

```python
marks={
    "math":80,
    "english":70,
    "physics":90
}
total = 0

for subject in marks:
    total += marks[subject]

average = total / len(marks)

print("Total Marks:", total)
print("Average Marks:", average)
```

Q7.HIGHEST SCORING STUDENT

```python
students={
    "Ali":85,
    "Sara":92,
    "Ahmed":78,
    "Zara":75,
}

top_student = ""
highest = 0

for name in students:
    if students[name] > highest:
        highest = students[name]
        top_student = name
print("Top Student:", top_student)
print("Marks:", highest)
```
